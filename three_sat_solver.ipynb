```python
#!/usr/bin/env python3

# Pradeep Sharma psharma8 

import sys 

# file_name = sys.argv[1]

# with open(file_name) as f:
#     inp = f.readline().strip()
#     inp2 = f.readline().strip()
#     part_assign = f.readline().strip()
#     inp3 = f.readline().strip()
# f.closed

### input 1
# inp = "-2,1;-4,2;-2,3;-1,-4;1,-3;-2,-2;-1,-3;3,3;4,-2;-1,-2;4,4;-2,4"
# inp2 = "-6,5,-9;1,6,8;-6,6,-3;2,8,7;2,6,-3;7,4,-4;-8,4,3;4,8,-3;3,-2;-4,-5,-3;-7,1,2;8,5,1;1,-2;-7,-6,3;-3,-1,2;-8,-1,-3;-4,-2;6,8,-6;3,6;-8,7,-5;-9,6;6,8,2;-4,4,-4;-8,-7;4,-6,-4;-1,7,-9;7,8,-4;-5,4,7"
# part_assign = "3=F,1=T"
# inp3 = "-1,-2,1;3,3,4;-3,-4,-1;1,4,-1;-2,2,-3;-4,-3,-2;5,3;3,3;-3,-4,2;1,-2,5;3,-3,2;5,2;1,-3,3;-3,-5,5"

### input 2
# inp = "-1,7;-6,1;1,1;5,2;4,-7;3,6;-2,1;-3,6;-5,4"
# inp2 = "6,-10,-5;-1,-8,-4;21,-11,-13;-2,15,-8;-13,9,17;-14,1,8;-5,20;-16,-20,-18;17,-21,18;12,20,-3;-2,-4;-6,16,-4;1,-20,5;9,-11,7;-1,9,-4;-2,15,4;5,-11,-10;-1,-10,2;17,8,-3;16,-1,-16;10,-12,-13;9,19,-1;2,-17;12,7,-3;-18,-12,-20;-19,-9,1;-10,11,-3;17,-1;8,14,-7;-15,-1,-2;9,11,-14;-16,4,-17"
# part_assign = "21=F,16=T,5=F,19=T"
# inp3 = "2,3,1;2,2,4;5,5;-2,-2,3;-5,3,-5;-3,4,3;-2,4,2;2,-4,-5;3,4,1;1,-3,-3;3,5,-1;-3,-1,-3;1,-1;-2,-2,2"

### input 3
# inp = "-3,-10;-10,-3;-4,7;10,8;4,-11;2,-3;-5,3;-8,6;5,-1;6,7;11,7;3,-2;4,-8;-9,-6;3,11;-4,8;1;-8,-7;10,-2;8,-11;10,-6;5,1;3,-4"
# inp2 = "3,-1,6;-4,4,-2;4,6,1;-1,-6,-3;1,-1,1;-2,-4,3;-6,6;-3,-4,-5;-6,3,5;1,-5,-4;-4,-5,3;-6,5,-5;-1,-2;-4,-4,-3;3,-4,-6;4,-5,-6;-1,2,1;2,2;6,6,-1;-3,1,-1;2,-3;3,6,-1;1,-3,1;2,-2,-5;-6,-5,2;-3,2;-2,-5,1;1,-3,-2;-6,2,-3;3,-1,-6;-2,-5,-6"
# part_assign = "3=F,5=F"
# inp3 = "-5,-3,-1;4,-2,3;3,-5,1;1,-5,3;-4,-5,-4;4,5,-1;5,4,3;-3,-5,-5;-2;-3,-3,1;4,2;5,4,-2;3,-1,2;-1,-3,-2;-2,1,2;4,2,-1;-5,5,4;1,-5,5;1,-2,-2"

### input 4
# inp = "-2,-3;-3,-5;1,1;-4,-1;-1,-2;-4,-3;3,1;5,3;-3,-4"
# inp2 = "-4,-2,-1;4,-3;-1,-3,4;3,-1,3;3,-3,-3;-4,1,-2;-3,-1,-3;1,-3;-1,-4,-3;4,-3,-2;1,-3,1;3,3;2,3,-2;2,3,-4"
# part_assign = "2=F"
# inp3 = "-1,-6,-2;-1,-3,-3;3;5,-1,-7;-2,-5,5;5,-6,-2;1,1,-3;-3,7,-7;-1,7,-3;-2,-3,6;7,-4,-6;-6,-3,-7;6,-2,7;6,1,-3"

### input 5
# inp = "7,17;4,-12;-1;18,-20;-30,6;9,-27;7,15;28,-16;7,-12;15,29;20,22;27,4;-9,27;-31,33;-10,2;-20,22;-29,-32;17,-12;-21,-15;-2,-2;-29,19;15,-22;24,-16;12,8;-34,4;-9,-19;20,33;29,-30;26,-23;-9,3;7,21;35,-8;-8;-23,-29;-19,-1;12,-22;8,-32;21,5;26,15;20;4,-1;-19,-33;6;29,14;-30;-13,11;-8,17;-7,-23;-25,33;-19,-5;10,25;22,-18;-26,13;-23,-18;-9,-8;-16,33;12,-30;29,16;35,10"
# inp2 = "3,6,-14;9,-6,-12;5,10,-7;12,6,-12;10,-13,13;1,1,9;5,10,1;8,-10;-10,-7;5,14,8;4,8,13;1,9,3;11,-13,-11;-11,-9,-11;2,-8,-8;-1,-11,-13;-10,-3,-7;-10,-1,6;-9,7,8;-7,13;5,-8,-13;-5,-8,4;-1,-14,13;-5,10,13;8,-8,-11;8,11,4;8,-13,14;7,10,-11;-5,14,14;1,-12,3;-13,4,-9;-13,2,3;-4,5,1;11,14,7;2,6,5;14,9,-9;4,-12,2;14,-8,8;5,-5,-14;-7,-11,-7"
# part_assign = "5=T"
# inp3 = "10,-11,-5;-7,2,-7;8,14;8,-10,-11;2,4,13;10,-3,9;-12,-1,7;11,4,-14;-11,11,13;14,1,8;14,-10,-5;3,-1,2;-4,-14;-12,-3,-2;6,6,3;-6,-4,6;-7,3,-6;-9,-9;5,-14;-7,-13,-3;2,1,-10;-10,8,4;10,4,-13;-6,12,-13;6,12,9;14,-2,-8;1,7,-12;-8,-7,-14;3,-10,10;-4,14,-1;7,-1,5;6,-7,5;-5,-8,-3;-12,-10;-1,-9,-1;-12,13,-7;14,13,-1;5,-13,13;-6,6,10;8,-1,4;14,-1,-6;12,2,-10;10,13,-14;-9,-3,6"

### input 6
# inp = "-18,21;-13,-10;2,-27;-24,-15;-4,-19;10,-21;-18,6;-32,-18;33,15;15,11;24,16;-2,-1;3,32;1,-32;-11,-19;-17,33;13,8;-14,-10;-25,7;-24,7;-5,24;5,-2;25,-23;27,-11;25,-16;14,6;22,-13;8,-29;-10,-30;5,-28;14,-16;-14,-18;6,3;-21,-26;2;31,-28;28,28;-26,14;33,-4;6,34;-8,25;-17,10;26,-8;-20,9;14,2;11,-17;10,-19;31,-27"
# inp2 = "-1,-4,-1;-7,6,-16;-9,8,-9;-1,-6,19;-20,19,17;3,-19,-14;-11,11;-11,-6,-15;-8,6,4;-9,-1,-7;10,-15,17;13,3,8;-14,-12,6;14,-7,-20;-6,5,22;16,-8;-11,-21,-21;24,12,-11;-20,-1,3;-3,-18,19;11,3,-19;-25,4;11,4,7;18,23,23;2,11,-4;21,-21,-20;17,-17,25;-4,21,-15;6,8,25;11,17,9;-20,-14,-25;-16,12,5;10,-21,21;-4,13;-6,10,17;24,6,14;-25,-1,-3;-5,-9,17;20,16,4;16,4,-8;2,13,14;-3,1,22;10,-9,-8;22,13,-19"
# part_assign = "24=T,6=T,9=F"
# inp3 = "5,1,-8;4,-6;-15,-2,-11;-10,-14,-15;1,9,9;9,-15,3;-2,5,1;12,-8,12;-9,-7,-15;13,-11,-3;3,-10,-8;-12,-3;-5,2,11;4,-1,-1;-4,-12,5;10,15,3;3,-11;-12,-11,7;-1,13,-10;2,-13,-7;-11,-15,5;2,-5;-9,9,-14;12,-15,-9;6,12,2;4,-10,-7;-8,2;-6,8,-11;-4,12,-3;-11,-15,-7;6,-11,-6;-10,15,-3;3,11,14;4,-12,11;3,10,-2;-15,-6,-3;15,2,-2;-2,8,11;4,-1;1,-8,-7;14,7,-10;-14,14,4;11,9,2;5,-15,11;13,-5,-14;-12,1,-9;4,15,-13;-4,10,-1;-13,14;-15,8,5;-9,13,-7;-9,-9,-1;11,11,12;-10,-9,-13;-1,-10;7,-7,8;-9,-13,-14;-13,14,-1;-10,-15"


### input 7
# inp = "-18,-12;2,11;13,16;17,8;28,32;-30,26;-25;9,-13;-5,6;-34,9;20,-3;-12,-33;21,-34;15,14;-2;28,-6;-16,-11;-19,13;4,-33;-16,-26;22,-28;-34,-16;-28,-2;24,-26;-2;-26,-8;16;7,-21;-11,-8;-27,-9;-32,-14;2,-22;14,-15;-12,-1;-14,24;-15,-9;18,21;-32,17;27,-30;25,-31;13,-24;10,-6;10,-9;-11,2;16,13;12,-29;29,16;-28"
# inp2 = "-4,-8,8;4,-9,11;-2,7,-4;-6,2;10,6,7;1,-3,-7;3,-9,7;2,-8,-11;11,-6,7;-6,-3,7;-10,-7,-11;11,-7,4;-7,-10,-8;-3;5,-4,-10;-2,-5,-1;3,-11,-9;8,6,-10;5,5,-11;-8,3,-6;-6,1,-3;8,2,4;-5,3,8;-1,1;-2,10,6;6,5,-11;-7,2,2;-10,4,-8;3,-1,-3;11,7;4,-6,-5;-3,4,11;3,-2,2;8,7,-1;11,9,4;6,-1,8;3,8,-6;10,5,6;2,4,11;10,-2,3;7,4,-6;5,1,9;3,-4,-8;5,-11,3;-1,-5;1,-3,-1"
# part_assign = "9=F,1=T"
# inp3 = "4,-16,-9;17,-11,-5;10,12,-22;-7,1,4;21,-11,2;15,-13,15;15,1;3,10,-1;-20,8,4;-15,-19,-17;10,16,10;-7,-21,14;-22,-7,12;-11,9,7;-12,-16,-14;-12,18,13;-19,12,-12;13,-21,22;-7,-5,-13;1,12,16;4,13,-15;-9,20,-17;13,7,12;3,-16,3;5,12,-15;-21,12,-3;1,-9,-11;-21,5,-15;-15,7,1;-11,19,-20;-18,-12,-15;-22,-17,15;-3,-7,-2;19,11,9;13,16,7;-5,8,-2;7,2,-21;-16,-21;21,-4,-8;-15,-17,-5;13,20,22;4,21,14;-1,10,6;8,-3;-21,-11,-11;8,-18,-15;-5,6;17,7,21;14,-5,3;-18,-12,18;-18,20;-9,-3;10,-22,13;21,18,-10;22,13,4;-10,13,-21;-13,-12,-3;-11,19,16;1,3,21;-21,15,6;18,17,15;2,-15;19,3,19;14,19,19;19,12,-22;3,-3,15;-18,-9,-20;-6,-21,-19;-19,10,11;14,-9,-16;6,21,22;10,-16,-22;15,21,-3;-5,7,-8;17,1,-14;20,-21,-17;18,-2,13;5,11,-8;-20,16"


## input 8
# inp = "-19,11;12,-22;14,-29;15,-17;-24,3;9;8,-27;13,-1;26,-5;1,-20;2,-24;-4,-9;7,28;1,-17;-27,16;-18,-18;-28,-6;7,-10;-11,16;27,-22;-5,-6;-10,9;-19,-30;16,30;-8,-25;26,31;21,9;24,17;-9,-24;-24,30;23,-16;-18,-26;8,-28;-22,16;-26,-22;-32,21;-27,18;-9;3,-14;-1,26;13,-20;30,15;11,25;14,15;17,16;27,32;4,-5;10,23;-3,4;-4,19;-20,-25;9,-7;1,9;25,13;-26,-14;20,5;-13,-15;28,-10;21,10;-31,-7;-4,3;-19,-3;28,32;9,-10;17,-29;5,19"
# inp2 = "-13,5,2;-2,-5,-25;-4,6,-26;-3,-17,-5;9,21,12;10,18,-16;-1,-2,11;-14,-1,9;-18,22,-25;15,-13,9;20,6,21;-4,8,26;26,-21,-13;-16,16,-14;-21,2,16;-17,5,25;-17,24,20;-26,-15,18;9,-25,4;16,12,3;-11,-26,-2;-22,-10,-23;-15,-7,-3;-6,14,5;-8,-18,-7;23,-24,-21;12,12,17;22,-4,-25;22,18,5;1,-25,-24;24,14,-1;-2,-2,19;-15,24,24;-18,12,-3;-16,6,26;-23,9,-21;23,14,-21;8,3,-26;-3,14,-3;-16,-6,3;-6,-1,2;-1,-8,-22;22,6,-2;22,26,13;-3,-5;6,-6;-19,1,-5;2,-22,-15;-20,24,15;10,-12,13;3,-20,1;-10,-11,-2;10,-18,16;3,7,-18;-6,19,4;-5,-24,-20;-9,-24,19;24,-14;-8,-5;-6,-9,-5;-9,-8,2;15,-25;-2,-12;-26,18,-9"
# part_assign = "4=T,9=F,17=F"
# inp3 = "5,-4,-14;11,-8,-19;-9,-10,-9;9,5,3;6,18,19;19,7,-4;-17,20,-17;9,18;14,-3,-15;-1,-14,-12;14,-3,17;-3,4,-9;10,20,-9;-12,4,-16;12,-7,-11;-9,11,-16;17,20;6,-16,-6;-10,12,6;-19,15,-17;17,9;-15,-17,5;8,4,3;-4,19,-12;19,-16,-19;15,-13,-17;-8,-11,-11;-20,-18,14;5,4,5;-6,-13,-18;-4,16,-16;-14,-12,10;3,2,19;1,-3,6;-15,-14,11;-5,-2,-12;-16,-9;-19,-2,6;17,-1,15;9,-7,12;-8,-15,18;-2,-9,-7;-17,-3,12;4,-19,-9;2,15,-19;3,19,-20;5,-10,-19;8,-7,-2;7,-12,-19;-13,-3,-16;13,-14,-9;-10,14,12;-5,13,-1;-14,-11,5;-12,-18,-14;1,18;-19,-13;4,-1,-20;5,-18,4;4,14,5;18,-5,12;4,-1,10"



## input 9
inp = "-77,-189;164,346;227,-96;367,222;-171,272;333,-341;328,229;-286,320;298,-106;1,350;283,145;-154,277;-161,370;-78,107;-28,-289;-275,378;148,-189;102,6;-155,296;-13,13;-34,187;50,179;165,400;266,-128;-169,-72;359,-179;223,-133;-259,165;80,-398;36,324;-51,-178;-329,318;-138,-231;75,-140;-111,-119;228,-40;218,-201;-52,-18;279,-179;9,21;379,-160;30,22;-84,120;318,-42;-149,1;139,-88;297,40;398,-162;-137,113;21,-48;137,73;273,186;-105,19;-219,398;-342,-310;-148,300;-266,3;-100,-247;-265,387;16,-239;34,-77;-359,-248;-252,-74;363,-314;-103,-229;-34,-276;116,-346;314,-201;152,-279;95,-154;-206,-107;204,344;162,100;43,-337;217,178;205,-158;-126,305;28,-15;-255,225;254,298;-133,383;387,-404;268,-48;313,92;-185,19;-147,180;-293,6;203,354;-176,-295;-349,214;-265,-394;-329,302;-8,-227;-158,-116;-368,86;318,-392;80,354;297,-26;-51,-274;287,260;-260,-2;-95,273;-108,185;348,59;53,-313;225;211,-269;32,43;309,-49;264,381;147,202;311,260;-30,-297;190,-281;286,-174;-45,-225;-300,-130;135,265;-68,269;233,139;280,-378;199,-344;62,357;-137,-98;53,154;158,-96;-80,-33;-81,-320;-46,186;-344,78;-34,-400;-378,-162;373,-52;-150,12;-136,225;234,-32;-302,-73;34,9;33,58;-92,17;-6,282;315,106;-315,-176;316,194;-403,-307;12,101;170,89;-116,365;63,264;-316,130;314,-399;176,-266;-72,-247;-95,170;186,160;-252,52;298,294;-185,-40;283,-317;-18,373;186,-371;356;-357,-110;-13,-26;389,220;-280,25;-314,180;-326,286;-315;-102,-132;-133,404;160,-394;235,-149;-34,-129;-191,92;3,-146;323,80;375,210;-294,-107;-77,-114;254;-126,372;-291,60;-42,-47;177,229;-297,-278;319,30;245,-2;-94,14;393,91;284,-118;-86,-40;-233,-274;162,-259;-348,189;121,-43;-389,-282;-11,-334;-321,50;-371,313;-403,-197;59,-260;-298,-235;66,217;346,331;-244,-282;129,-260;-155,-310;-237,-362;-238,58;-248,-388;42,153;332,-209;99,52;301,-208;-350,6;-217,359;230;385,-3;-64,-82;-121,219;-327;-126,-234;-371;-46,365;121,-252;-388,9;400,-295;-326,-234;335,-246;-254,376;-177,82;-365,-41;-296,-304;-381,-154;-382,29;-332,-265;124,195;161,259;189,402;266,89;-264,76;-40,-189;-204,404;332,-332;28,-404;224,230;49,-346;-246,-54;-18,56;-400,165;-182,-157;-9,72;-131,-86;-314,-280;-239,-328;-269,-403;-278;-377,213;-322,288;184,244;-159,69;-122,-305;-242,325;-103,27;363,-285;390,75;-283,392;155,-128;106,-86;338,215;-266,288;-395,254;-313,318;319,-16;45,284;-340,328;294,364;74,384;348,-335;212,152;-235,-147;-249,325;-284,-276;378,9;-187,-375;322,-85;-253,-41;53,-113;-348;-15,361;-320,-127;-297,117;187,230;404;-164,165;-133,-219;79,-121;-161,146;43,-334;-34,400;-235,-404;-259,-229;-308,-119;88,386;-350,228;69,-259;362,-148;171,85;72,-347;-234,56;-91,-244;141,95;-199;357,234;59,189;-68,138;404,-376;270,-104;-164,335;-247,-401;-154,-333;309,105;-248;-24,397;9,359;-291,-334;204,-400;-303,-83;-375,83;-386,400;135;-112,378;-173,-199;171,-266;181,181;60,135;232,224;-187,-266;-161;237,-50;-77,-67;11,-175;-120,56;223,-217;3,-93;-91,-44;151,-377;60,-135;-178;-334,175;21,-132;-99,231;225,-156;101;378,218;-78,-122;316,309;-145,145;-310,52;37,146;36,-115;-128,-329;-46,-155;-50;21,-339;-202,163;164,354;262,-43;-369,358;247,-52;318,-68;-154,359;367,-353;390,120;-264,249;-198,244;-369,135;-377;-172,-328;120,58;27,-29;103,-97;-280,347;193,161;-360,-114;176,-233;209,-156;-213,-128;166,-264;359,-382;-204,253;-91,392;394,404;-300,43;-55,59;31,-154;359,-189;14,154;165,-270;359,-200;151,226;33,-364;-361,300;132,160;-348,255;-61,236;-148,339;-333,388;-43,-194;-156,134;-245,263;-70,140;310,-103;135,-99;-319,284;-13,303;135,400;384,327;-356,6;224,308;-267,-257;-319,31;-265,69;279,-92;-26,309;215,71;-382;-363;-305,251;390,-142;238,133;-116,-327;-56,-249;-380,213;-302,-200;-353,96;369,-389;-400,-355;-107,121;-227,-262;346,301;-72,-213;259,318;121,-167;-65,343;373,-164;-259,115;251,350;342,-173;-229,291;62,-266;-71,159;91,86;-193,42;145;-45;-195,191;-252,95;-192,163;140,-101;-119,163;-279,-68;-276,-305;89,-376;31,-339;35,380;-74,-391;23,309;24,106;-92,-134;295,-192;-340,-53;-160,103;116,-307;-232,305;-145,171;168,364;-338,-64;-361,-295;-255,-242;26,-364;-367,183;-135,-156;197,-360;158,123;-141,-131;-255,-102;300,-229;228,64;366,-327;231,182;-97,-63;-305,368;63,262;104,185;-376,-353;-364,52;259,-283;227,298"
inp2 = "62,-16,-54;72,4,57;4,-49,-32;-56,-88;-29,57,-34;-87,-77,59;84,-8,-39;66,3,-86;74,-62,-21;43,44,60;17,65,84;-16,-73,-26;22,-30;27,33,50;-20,-55,51;-19,17,-85;-4,67,20;66,-41,-73;29,2,34;-75,-12,4;-7,-33,53;-86,3,-55;32,7,58;88,45,-64;-67,49,3;36,15,39;-83,-30,69;61,9,55;26,-19,52;-74,-31,21;-6,85,-76;-2,-43,-46;-47,-37,-14;-50,-44,10;-68,18,-12;6,-69,-18;86,70,69;82,50,80;29,-14;30,12,-7;77,42,-78;16,31,31;-42,-5,34;36,86,45;21,-53;-3,-67,40;-77,-25,-57;67,-40,69;59,88,52;-12,-6,-57;28,-62,-51;9,45,87;58,-76,-26;-32,37,40;60,57,41;52,-17,61;-54,82;-76,-40,10;-32,54,60;-17,62;-29,-3;48,-8,-88;-30,-77,47;84,-12,-79;54,-33,-84;-28,-42,-46;-5,60,82;-76,-47,9;-85,79,-37;7,5,-65;-54,46,20;-88,-8,34;-22,-32,-77;77,-73,-86;-16,-26,60;44,5,-31;-71,-42,-37;51,-69,-29;-11,-2,47;-53,-36,83;-28,-3,48;-46,84,25;-36,-69,-53;30,-7,85;-20,58,10;8,50,-33;19,46,-28;15,-65,41;67,31,59;72,-13,-68;5,84,-10;-32,63,-20;-66,-77,4;-23,10,62;86,9,72;34,20,-25;-52,73,23;52,-52,-18;-80,79,-19;-48,55,-26;52,22,43;14,-73,19;77,-70,85;54,-86,69;48,60,-81;-41,72,-15;15,71,78;-55,-70,36;-70,39;-4,39,20;-24,-29,-30;-73,26,80;39,52,3;-61,-26,-83;72,-85,-65;-26,-87,-41;26,11,71;-46,-28,6;-73,-22,-28;61,-59,-32;8,73,-46;-14,-21,85;39,-55,50;-45,-63,74;-6,-33,68;-9,-5,78;24,25,-47;-83,37,-32;-81,-7,-33;24,-32,-72;-87,47,-40;-48,67,-56;-72,-75,-81;-26,7,-73;3,64,78;1,-28,10;-14,42,-59;-72,71,-38;-63,-57,-65;43,-20,-35;-34,68,62;-36,88,9;-77,-33,26;78,-83,-17;-23,-14,-87;64,-10,6;21,39,-53;35,14,-61;17,88,36;-32,-50,-39;28,-64,-60;-75,-55,-20;46,-71,-15;-29,68,59;5,7,-64;-15,11,23;-46,37,-88;-39,-34,-48;-23,65,74;21,-3;51,28,-14;14,83,68;13,59,-4;-36,55,23;-43,50,-82;-23,45,-86;-45,73,-84;-85,-57,10;-35,-40,-15;12,-58,71;-22,-31,18;-31,-47,-6;78,28,4;-42,-67,73;-18,31,30;59,19,29;19,-52;52,-36;-80,-49;83,59,-20;-59,5,-82;34,-3,-49;25,42,87;-25,-20,56;-73,-17,-34;-49,-58,80;-64,54,8;-77,-23,58;33,4,-57;-77,67,-83;-47,87,71;17,41;-48,-48,13;-66,-68,81;-77,10,-50;-39,62,-60;-83,19,62;60,-11,81;84,-71,6;-2,54,11;-12,83,61;-33,-3,-37;68,10,75;-23,29,11;-75,-8;13,-5,-10;48,72,32;-30,5,4;-71,49,-7;-55,-72;-16,-31,-67;-7,-10,25;-17,80,-68;-47,-30,-10;-86,-64,-72;-53,29;-11,-43,65;85,-17,-74;-2,16,72;29,1,57;-60,84,-36;62,73;37,49,-87;55,23,-84;-50,14,78;65,10,-30;-34,78,30;29,6;78,-14,-43;59,73,-82;64,10,-45;-73,-29,-61;-25,88,-15;68,-38,46;39,62,81;55,-56;75,37,48;58,34,-57;13,69,-10;3,85,13;31,70,38;5,-20,9;48,50,77;27,-55,-85;82,-71,-18;54,58,-32;-34,-14,84;-29,36,-87;18,-77,54;-76,61,-52;63,-14,17;84,-75,-63;80,80,80;-38,58,-18;4,7,36;-57,82;-62,-32,77;85,12,-56;37,80,-26;-17,-44,1;85,-33;39,2,2;-41,-39,44;84,27,-17;-80,-53;-43,24,-65;33,59,-35;-14,58,-26;59,38,-60;50,3,-10;-84,-57;-45,-41,78;65,39,21;28,-55,31;64,-38,3;12,-51,60;-38,30,-7;-88,78,61;63,-43,10;12,72,-75;86,44,2;-37,-53,-60;-24,-79,-56;72,67,20;47,49,70;28,55,-85;-10,-4,27;-49,-13,-71;-6,-59,57;-21,52,37;-41,20,-83;-60,-39,64;5,-50,4;-72,60;14,-60,-47;54,-43,25;1,37;-53,-28;-9,-59;-71,-72,26"
part_assign = "54=T,4=T,63=F"
inp3 = "-57,-87;64,-73,-50;19,78,37;-85,35,59;-50,-18,72;37,-68;-22,-68;-2,-96,90;-54,-85,-63;24,24,6;11,-10,34;26,-56,-52;-9,-26,23;-19,77,46;23,60,-26;27,-57,-8;-94,-91,-88;65,-9,-74;32,97,89;53,-91,-17;-2,87,-3;55,-48,39;46,-21,-81;-43,-36,-28;-9,-11,20;93,-10,88;74,40,-50;-6,7,-53;-41,-12,-60;-36,-2,-94;38,58,-89;74,-15,-37;-71,-9,8;63,-21,-32;84,36,11;-31,-16,-57;17,-6,-60;-71,-59,-72;-21,63,-48;84,-2,94;54,77,-63;63,23,95;74,9,12;-52,-37,-91;42,-43;16,59,-53;-54,-43,-58;66,59,-45;11,82,-76;-51,-50,-72;-75,-20,-30;-6,75,86;38,36,-75;91,89;-17,-8,34;6,6,93;-63,50,-40;41,-58,1;-77,22,35;3,-73,47;50,-82,27;-78,-92,94;31,77,61;28,47,-26;-53,-27;4,28,96;-83,33,10;-60,-42,38;-70,-96,-58;-30,73,-66;-53,-4,-43;-76,-56;-43,-10,16;59,38,-12;-62,-45,60;50,-48,11;65,-55,13;30,-6,2;-42,11,-90;-88,-69,63;72,61,-97;7,-4,41;-20,-15,-71;35,12,72;-59,-41;47,-12,84;-2,41,31;86,49,-4;67,-30,73;84,42,-80;-8,-23,61;41,75,-16;-7,-88,54;-2,-70,-64;-83,93,78;-7,23,29;-86,-54,-58;-37,27,87;24,-86,45;-77,95;-92,-59,-61;-56,-33,-87;38,-64,-62;42,-30,40;-62,-56,-80;5,24,65;17,6,46;-19,-97,55;-14,-8,-26;-34,-60,-50;-43,60;53,81,-36;78,13,-70;-96,-30,-16;-30,-86,56;9,79,81;25,-44,67;-68,-22,-59;-5,-80,-74;40,8,90;62,-54,6;9,-23,72;-65,-94,-20;-24,-40,-72;-89,-3,83;-80,79,-45;25,22,43;89,78,34;86,39,22;-31,-72,-71;67,41,-42;70,8,12;-9,24;-96,7,34;64,-46;-57,-49;-78,-46,21;70,90;25,-63,-38;69,-57,18;89,35,46;45,70;8,-40,21;74,54,-70;22,84,-6;-48,7,9;74,-50;-43,28,-7;-46,28,66;-33,39,65;-13,-79,-65;-6,-48,1;83,9,67;71;-97,91,-55;-87,40,-23;-87,86,-44;-4,-39;-82,-11,77;76,21,52;8,9,-84;-9,76,12;-84,-35,57;41,39,-75;29,-17,-17;-36,-24,95;63,78,-47;-83,16,-56;-19,8,9;64,16,35;-59,38,-79;-7,-17,48;57,13,38;-85,-82,-31;-63,66,-83;31,-81,79;-48,-77,-63;-33,-14,91;-54,-41,-31;27,22,22;-56,18,75;30,88,42;96,-53,-18;-81,-35,39;-75,13,81;-43,12,27;74,73,-13;96,76,-27;69,95,-31;-75,24,-12;29,-82,84;-64,-1,-43;41,-95,-48;69,-73,13;22,18,-46;-70,9,73;-16,72,-37;64,91,-20;-69,-64,-39;-56,-66,16;-39,-63,42;-92,37,-46;42,92,-37;48,40,-75;30,-78,7;72,-5;-33,60,-18;-82,77,66;91,-60,-85;49,97,92;48,-29,18;-74,-28,-39;54,14,68;-89,38,80;-67,-4,-37;19,-34,-13;-96,-97,-56;-78,-4;-60,15,74;-83,-7,-25;-50,-10,-70;-41,12;1,-95,10;56,74,-88;-41,70,86;-15,-88,59;-61,-89,41;-96,79,-56;-53,96,-74;69,21,-47;45,-50,57;-24,-44,93;33,59,41;-95,-45,-84;78,33,-95;-73,86,-7;86,72,27;-91,-48,91;17,-94,67;-71,-96;-5,94;47,-77,-29;5,32;6,-61;-32,-73;82,76,42;-18,19,69;38,-21,18;60,-7,-97;89,-72,85"

### input 10
# inp = "-48,2;32,47;17,-44;-63,16;-56,-48;57,-46;15,11;57,-70;-59,-64;-13,-12;-32,-61;-13,-38;-41,17;-61,-40;4,-26;-4,37;-21,-33;28,-43;66,-50;60,52;-70,-18;-52,5;-63,-36;14,60;-24,-62;-35,30;-31,-33;22,-10;9,-2;-44,62;-18,42;-12,-53;-48,-3;-40,69;-55,-22;-29,-28;47,-56;35,61;-60,41;-12,-63;47,52;57,71;65,59;40,64;-45,63;-38,56;-26,61;-56;-69,60;-41,8;-58;-4,21;-6,-37;-15,-25;62,-65;-51,-35;55;-26,-69;40,36;-5,25;-59,-49;-10,-24;-47,-25;3,8;-21,-56;2,-7;-41,52;-42,-60;40,-36;67,-37;-12,10;53,11;-43,-71;69,-31;-49,-44;-46,-12;17,38;29,-7;56,-41;-9,65;15,-1;-63,59;11,-65;56,-69;-36,-60;14,-37;43,-12;53,-27;10,25;29,-63;-11,-52;71,33;18,-57;2,-68;12;24,-32;51,53;67,-13;10,29;53,22;28;22,66;50,-55;-18,-20;37,17;23,3;20,-65;54,-42;61,70;6,64;40,43;-23,3;-13,71;67,-33;55,11;-36,-65;63,4;65,-39;50,-51;-54,12;-16,4;-18,18;54,-1;54,-71;13,1;65,9;26,12;69,-26;-36,11;-29;-47,28;53,-63;-59,-35;48,-36;33,28;25,51;66,-60;-54,10;-9,38;13,5;-46,21;-35,-50;13,36;20,-70;-66,28;-39,61;-69,-42;21,-64;43,-47;55,-8;-54,3;9,-26;20,55;-8,28;10,1;-43,-26;-36,36;-62,-19;-38;-41,-44;66,28;43,16;31,-19;14,11;-23,-71;-24,24;54,-28;-5,63;-62,-60;60,-50;-40,56;33,69;-14,16;-32,-45;-47,43;-43,-22;35,-61;-70,48;-19,-36;3,11;3,57;-7,14;-57,62;-51,24;-15,-54;42,62;-36,7;32,-4;-45,22;-65,-1;-43,5;-5,62;-23,-48;-20,-69;-43,46;-27,-65;-1,48;-24,35;-45,-61;22,40;36,-19;24,-26;48,-61;62,14;-8,40;51,-32;45,14;-56,26;-59,2;57,-54;48,-41;26,5;61,24;43,-61;6,31;-63,69;32,71;8,-33;12,38;-28,-11;46,-70;10,-19;20,10;-4,-38;-40,-70;19,-10;56,-51;53,-37;-46,60;49,-42;-61,-1;22,48;-12,-12;-21,-60;-60,-67;2,-46;-41,62;-33,-40;-14,-21;53,46;14,57;54,-55;4,47;33,30;-53,11;62,47;-25,-27;21,-31;-56,59;66,42;37,25;43,-19;-19,-26;44,31;23,43;17,-71;-38,-4;-46,66;18,11;-61,-55;19,-33;5,67;28,-63;-39,-46;16,49;-15,39;48,58;28,23;47,-22;-3,24;30,66;-23,36;59,-44;-53,-25;-5,-56;-63,-36;-8,6;53,15;-33,-35;-62,32;29,-8;-28,70;19,17;35,-54;10,-13;-13,-18;-49,-57;7,-71;-66,-71;-63,1;31,-34;59,58;34,35;37,40;62,-67;-50,-9;55,65;-25,67;-71,-22;67,-7;-9,-9;54,-21;-2,-8;59,-61;24,-45;53,-1;59,-55;30,-52;7,43;44,36;7,-65;-45,-48;-70,10;-33,-11;-12,54;-20,62;21,46;38,59;37,-3;-8,-15;69,-28;65,5;51,-63;58,63;-53,15;2,-21;10,-53;-19,-28;-16,-36;30,3;60,55;-13,70;-42,15;5,-11;62,-66;64,44;-38,-21;-9,-64;34,8;15,-22;43,12;-10;6,20;44,-42;35,-66;19,-31;34,29;22,21;65,-66;-51,-5;-61,48;-52,-15;4,-8;-52,-40;19;24,49;40,29;-58,-20;6,59;43,41;-35,-67;26,22;-38,-17;10,-67;67,59;-21,69;61,10;37,44;-12,-52;19,52;-18,-2;41,8;2,-69;-23,-2;-31,-4;69,12;22,-39;65,43;19,21;67,4;-8,-45;-31,6;69,46;16,64;-61,35;8;23,-61;-17,55;-34,70;69,11;-26,9;24,48;63,10;53,-31;4,52;67,-17;-46,18;-59,7;-17,64;44,-41;-60,39;-11,-12;14,46;-31,45;-69,-30;-68,56;64,9;-39,22;58,60;-43,33;-29,69;-36,-53;-23,29;6,-17;70,-12;-19,-32;32,-67;5,-68;45,-54;48,-35;14,67;48,71;6,-27;-15,42;69,-70;-59,-38;34,59;-29,-61;-30,48;-17,54;1,-61;64,-36;-37,-32;-35,-33;-67,-64;69,-48;-57,-14;-21,61;-31,62;-27,34;-42,60;-28,-38;25,-58;-55,-56;68,39;-2,5;-8,-15;-19,38;-52,22;-9,70;-64,54;35,6;-47,-21;17,-41;-68,-11;-41,41;11,54;-27,6;-60,-27;43,-56;-65,-19;-31,37;62,29;-26,-56;-15,50;-34,64;-9,66;19,-29;17,-5;-26,-20;-20,3;-6,67;12,-71;5,-61;13,25;-41,1;-16,71;41,59;9,15;-10,-28;-43,57;-5,-19;70,19;21,-59;-32,62;-62,-67;1,64;-18,-67;13,13;-68,-28;-9;-43,-63;-10,-4;-3,47;-15;-56;1,-62;35,-58;-34,-40;-42,-39"
# inp2 = "13,-18,26;10,-13,-11;-19,-17,3;-1,16,17;-22,4,20;-20;20,6,-18;4,11,4;23,-16,2;-25,-8,14;-26,-6,10;25,-9,25;-14,25,14;-15,-5,-10;-26,19,11;7,-18,15;-17,20,-1;18,-15,-14;-19,-12,-14;-4,12,-16;20,-26,19;1,6,3;-8,7,26;18,-16;26,-8,2;-21,26,-18;-18,16,-17;10,-24,16;-15,22,12;-19,26,-6;-22,-19,1;16,-25,8;-20,-12,-11;-12,-24,25;-1,-16,1;21,-20,-22;-21,8,-21;-17,14,-11;-9,-17,16;-1,-5;10,25,16;25,24,16;21,21,-3;22,-17,3;-2,-24,-19;-3,-10,15;-26,-24,-12;3,12,3;-15,-1,-15;14,1,20;-16,18,20;13,21,10;7,26,-26;-10,-3,-12;20,-15,24;-13,-11,-3;-21,25,-18;-14,-15,10;-11,-3,26;11,8,-5;-15,-21,2;16,13,17;-12,-26,-6;-14,-2,21;9,-18,16;13,-23,3;-6,-10,-22;19,10,7;-23,-11,-20;10,-12,-7;-20,-25,1;-16,-22,16;-2,7,-26;-17,-4;12,12,-19;21,-6,-9;-24,9,-14;-22,18;18,2,1;19,26,-21;11,17,-2;-9,-12,12;26,-26,-2;-3,-19,-21;1,-5,-6;-16,-4,20;11,16,26;26,15,-19;15,5;25,22,26;13,5,-2;-4,-16,15;-15,10,-5;9,17,-8;10,-14,14;-17,-20;-11,22,14;7,-6,-8;-2,-24,22;10,1,-26;-5,3,2;15,12,-8;3,22,-5;9,-23,12;19,-2,11;16,-6,-11;-17,-7,11;10,22,-5;-24,20,-5;5,-13,18;-21;16,10,15;10,19,-22;14,-9,24;-19,-1;25,-9,-16;-21,18,23;-2,-26,5;-22,4,10;-23,18,1;-11,18,22;1,14,-1;-23,9,-26;15,14,-13;-12,2,-13;20,-12,11;-3,17,-10;-14,-25,-9;5,-14,20;-14,9,-21;-3,3,-9;18,-10,12;-6,-1,2;-17,3,25;10,16;-4,3,20;-10,-14,-11;-9,23,22;-14,18,24;3,9,-3;9,-6,-23;-10,-3;-26,-24,-25;-19,-12,-22;8,10,23;19,2,20;21,10,24;-3,2,-6;26,17,-25;7,17,-14;3,-15,20;16,-9,7;-15,-9;4,6,-14;16,-5,-24;-24,5,-3;-14,-9;-16,24;-21,-16,19;11,-9,-10;-8,-24,5;-1,5;17,18,-3;-21,17,6;13,4,2;-23,-3,17;8,9;-2,18,-16;13,-13,8;-11,11,6;16,-8,-5;-10,25,9;8;1,-19,-16;26,13,-2;20,-11,-14;-26,-24,-12;11,-22,18;-8,-13,3;13,-26;-11,-2,-7;-25,17,-7;7,-20,-13;-20,-4,5;1,7,17;12,-1,-21;-15,-26,-21;-24,-1,-15;1,-7,-26;-3,-14,16;16,10,-7;20,-25,-4;21,-25;-22,21,-19;-19,-23,-10;-22,2;20,-19,-18;6,-13,4;7,8,-6;1,5,-7;-26,5,-11;22;18,-5,12;-19,16;4,-10,10;5,6;-24,13,13;12,10,-8;17,1,-4;1,-13,13;25,-19,23;-25,-1,7;-17,12,-7;17,23,-2;-10,12,-4;12,6,-17;16,-25,-2;19,24,-13;-24,-12,-19;-11,7,1;-3,-9,-24;2,-6,-14;-13,-21,-4;-4,-24,7;25,-25,-19;20,5,3;-8,-23,2;8,-18,-11;-2,19,14;8,3,-3;8,11,10;20,22,17;-7,-25,5;-3,-20,9;8,-2,2;5,2,-6;-8,26,-26;-24,-8,-21;5,-7,-12;5,4,-6;17,19;9,16,9;23,3,-5;10,10,8;7,-12,20;14,25,-4;-21,1,-6;-17,-5,3;-1,6,-6;-3,-10,26;8,-15,-9;-6,19,-12;-6;-1,-24,-14;3,-2,-8;11,-21,-18;-11,6,4;11,19,-4;-12,-19,-9;2,10,-26;-23,-19,-8;-14,-11,-23;24,-4,-18;-4,-16,17;18,21,-3;7,-22,22;21,21,12;-3,-7,5;26,-14,1;-21,24,26;-14,-11,-10;-12,-2,22;13,-15,24;-17,-6,18;-16,-1,-16;-13,-19,26;-16,3,-11;24,1,1;23,-1,4;19,-12,-18;16,23;15,21,5;-19,-23,18;20,3,-1;-26,-2,24;-22,-21,5;-1,-19,3;-4,6;12,1,26;-7,11,-22;21,-8,3;-5,-12,11;24,-6,21;22,21,-9;15,2,5;15,-15,-24;13,17,4;4,6,3;-20,21,25;17,26,21"
# part_assign = "14=F"
# inp3 = "-71,-49,-38;-78,36,4;-65,5,-6;-6,1,61;-76;-17,60,56;-77,-64;24,-16,-12;-60,-28,-56;38,-65,51;58,37,-6;54,68,-3;72,-11,64;8,-36,-11;-22,59;-17,49,-57;58,-28,-27;63,29,18;-45,48,-28;71,-39,11;35,37,-37;-49,65,-46;-85,51,-9;-3,-61;65,3,71;-47,-27;-2,15,-72;49,50,-80;-72,74,-11;26,8;15,10,2;51,-5,-65;-59,76,6;-19,-15,15;-56,18,-79;49,-50,22;-2,39,-3;-60,15,3;85,15,-84;-33,79,-1;26,-72,31;3,54,2;66,-60,-13;63,17;7,5,82;-55,-35,32;77,-55,66;30,3,-29;36,59,-50;33,-15;81,28,68;32,39,63;-51,-6,24;-41,31,5;-47,65,-69;-69,29,-50;81,33,19;-44,-25,28;-16,13,10;51,41,-1;-68,-24,68;12,76,29;-70,27,-63;-1,26,-49;-46,36,39;-58,-20,82;23,42,73;-23,12,-40;-38,-10,62;-65,48,33;40,-41,67;-49,43,18;38,-23;-73,76,-7;20,-35,-60;35;13,30,52;46,35,-44;44,59,-73;77,61;-41,-75,28;43,-62,-84;38,22,18;-45,-31,-61;55,70,-58;40,-19,-11;73,6,-27;-47,13,76;-31,-33,72;76,-51,-4;13,78,-55;22,85,-72;-59,34,73;-13,29,-76;23,34,32;8,2,76;71,-78,-61;-77,75,19;21,-70,1;55,15,33;-51,72,79;-83,65,-27;-9,-3,-62;55,-29,50;9,-29,-70;21,25,-17;54,41,-18;-79,2,61;57,-61,14;66,-80,-84;-22,-79,24;54,50,32;-40,6,-30;-22,-80,2;74,-48,-38;-16,-72,-6;55,-46;72,26;52,33,-9;-56,-38,70;60,64,50;-64,59,-19;50,58,80;-28,18,84;-9,-61,-6;-6,-83,-1;-8,28;44,-25,-61;-8,24,-45;-8,81,21;37,70,79;-46,-14,-57;81,71,24;-7,-35,2;-33,47;-76,20,79;37,-10,36;-68,-67,20;67,-85,-56;79,23,83;-73,5,69;-56,-79,-39;-19,-14,1;34,32,-57;-52,25,77;69,-7,-26;-80,36,-36;-2,-11,-35;23,-10,-82;-22,-9,22;62,18,63;61,75,-39;-26;51,58,-56;-18,12;-69,-24,-43;44,78,77;70,-24,-41;-10,-54,-71;82,-7,-20;67,9,-11;24,75,-13;-39,43,-46;9,-31,31;-77,12,-28;47,-82,52;-77,-24,64;-78,38;60,34;-78,-49,16;-25,13;76,55,83;-75,18,-22;58,70,-29;-16,-49,-65;-37,-54,52;-11,31,-13;-70,21;24,-7,-22;-58;-84,-59,37;-45,-20,46;-32,82,45;-74;-24,80,34;-5,-48;-25,-15,-44;78,-6,76;66,73,-4;-65,-81,28;-83,-46,-40;-13,-16;9,-74,-14;-63,3;-73,-39,51;-10,-2,12;-53,-19,-67;-85,24,30;-43,-29,-65;-64,31,42;-9,-51;40,-53,-84;-23,47,-25;-10,-65,-71;7,85;28,-5,33;27,-38,21;19,-51,-34;-16,17,67;74,-17,-27;-81,-18,-13;-28,-33,-58;7,-67,-14;22,-72,78;9,-23,-27;-26,-9,-30;49,40,-48;62,-82,41;-66,-48,64;30,21,-20;-2,3,-1;-38,46,-42;-27,-67,20;75;62,-71,-10;80,1,-73;-56,62;56,4,-63;-70,-40,-49;65,2,-50;-77,-71,-6;-11,63,45;-81,77,44;-62,40,4;-18,37,-59;-22,77;53,-10,24;-3,32,79;61,-28,-14;19,64;-40,-36,-40;-81,18,10;60,78,68;-30,-28;53,-2,-10;-1,-28,-73;-82,-77,22;-40,67,-44;-25,-54,48;73,-35,50"




# # # # # # # # #
# # Page One # #
# # # # # # # #

def rec_7_alg(clause_list,var_dict):

    if(len(clause_list) == 0):
        return True
    # Base case , only one clause in formula
    if(len(clause_list) == 1):
        clause = clause_list[0]
        
        if(len(clause) == 2):
            [a,b] = clause
            if(a[0] == '-'):
                a = a[1:]
                a_bool = False
            else:
                a_bool = True
                
            if(b[0] == '-'):
                b = b[1:]
                b_bool = False
            else:
                b_bool = True
                
            assigned_a = (a in var_dict)
            assigned_b = (b in var_dict)
            
            if(assigned_a and assigned_b):
                a_or_b = ((var_dict[a] == a_bool) or (var_dict[b] == b_bool))
                if(a_or_b):
                    return True
                else:
                    return False
            elif(assigned_a):
                if(var_dict[a] == a_bool):
                    return True
                else:
                    var_dict[b] = b_bool
                    return True
            elif(assigned_b):
                if(var_dict[b] == b_bool):
                    return True
                else:
                    var_dict[a] = a_bool
                    return True
            else:
                var_dict[a] = a_bool   # assign bool value to either a or b to make the current clause True
                return True
        # # Base case for caluse with three variables
        if(len(clause) == 3):
            [a,b,c] = clause
            
            if(a[0] == '-'):
                a = a[1:]
                a_bool = False
            else:
                a_bool = True
                
            if(b[0] == '-'):
                b = b[1:]
                b_bool = False
            else:
                b_bool = True
                
            if(c[0] == '-'):
                c = c[1:]
                c_bool = False
            else:
                c_bool = True
            
            assigned_a = (a in var_dict)
            assigned_b = (b in var_dict)
            assigned_c = (c in var_dict)
            
            if(assigned_a and assigned_b and assigned_c):
                a_or_b_or_c = ((var_dict[a] == a_bool) or (var_dict[b] == b_bool) or (var_dict[c] == c_bool))
                if(a_or_b_or_c):
                    return True
                else:
                    return False
            elif(assigned_a and assigned_b):
                a_or_b = ((var_dict[a] == a_bool) or (var_dict[b] == b_bool))
                if(a_or_b):
                    return True
                else:
                    var_dict[c] = c_bool
                    return True
            elif(assigned_a and assigned_c):
                a_or_c = ((var_dict[a] == a_bool) or (var_dict[c] == c_bool))
                if(a_or_c):
                    return True
                else:
                    var_dict[b] = b_bool
                    return True
            elif(assigned_b and assigned_c):
                b_or_c = ((var_dict[b] == b_bool) or (var_dict[c] == c_bool))
                if(b_or_c):
                    return True
                else:
                    var_dict[a] = a_bool
                    return True
            elif(assigned_a):
                if(var_dict[a] == a_bool):
                    return True
                else:
                    var_dict[b] = b_bool
                    return True
            elif(assigned_b):
                if(var_dict[b] == b_bool):
                    return True
                else:
                    var_dict[c] = c_bool
                    return True
            elif(assigned_c):
                if(var_dict[c] == c_bool):
                    return True
                else:
                    var_dict[a] = a_bool
                    return True
            else:
                var_dict[a] = a_bool
                return True
            
            
    # # if there is more than one clause in formula       
    else:
        clause = clause_list[0]
        clause_list = clause_list[1:]
        
        if(len(clause) == 2):
            [a,b] = clause 
            if(a[0] == '-'):
                a = a[1:]
                a_bool = False
            else:
                a_bool = True
                
            if(b[0] == '-'):
                b = b[1:]
                b_bool = False
            else:
                b_bool = True
                
            assigned_a = (a in var_dict)
            assigned_b = (b in var_dict)
            
            if(assigned_a and assigned_b):
                a_or_b = ((var_dict[a] == a_bool) or (var_dict[b] == b_bool))
                if(a_or_b):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return bv
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    clause_list = [clause] + clause_list
                    return False
            elif(assigned_a):
                if(var_dict[a] == a_bool):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    var_dict[b] = b_bool
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        var_dict.pop(b)
                        return False
                    
            elif(assigned_b):
                if(var_dict[b] == b_bool):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    var_dict[a] = a_bool
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        var_dict.pop(a)
                        clause_list = [clause] + clause_list
                        return False
            else:
                # # T T
                var_dict[a] = a_bool
                var_dict[b] = b_bool
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # T F
                var_dict[a] = a_bool
                var_dict[b] = (not b_bool)
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # F T
                var_dict[a] = (not a_bool)
                var_dict[b] = b_bool
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                else:
                    var_dict.pop(a)
                    var_dict.pop(b)
                    clause_list = [clause] + clause_list
                    return False
            
        if(len(clause) == 3):
            [a,b,c] = clause
            if(a[0] == '-'):
                a = a[1:]
                a_bool = False
            else:
                a_bool = True
                
            if(b[0] == '-'):
                b = b[1:]
                b_bool = False
            else:
                b_bool = True
                
            if(c[0] == '-'):
                c = c[1:]
                c_bool = False
            else:
                c_bool = True
            
            assigned_a = (a in var_dict)
            assigned_b = (b in var_dict)
            assigned_c = (c in var_dict)
            
            if(assigned_a and assigned_b and assigned_c):
                a_or_b_or_c = ((var_dict[a] == a_bool) or (var_dict[b] == b_bool) or (var_dict[c] == c_bool))
                if(a_or_b_or_c):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    clause_list = [clause] + clause_list
                    return False
            elif(assigned_a and assigned_b):
                a_or_b = ((var_dict[a] == a_bool) or (var_dict[b] == b_bool))
                if(a_or_b):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    var_dict[c] = c_bool
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    else:
                        var_dict.pop(c)
                        clause_list = [clause] + clause_list
                        return False
                        
            elif(assigned_a and assigned_c):
                a_or_c = ((var_dict[a] == a_bool) or (var_dict[c] == c_bool))
                if(a_or_c):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    var_dict[b] = b_bool
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        var_dict.pop(b)
                        clause_list = [clause] + clause_list
                        return False
            elif(assigned_b and assigned_c):
                b_or_c = ((var_dict[b] == b_bool) or (var_dict[c] == c_bool))
                if(b_or_c):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    var_dict[a] = a_bool
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        var_dict.pop(a)
                        clause_list = [clause] + clause_list
                        return False
            elif(assigned_a):
                if(var_dict[a] == a_bool):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    # # F T T
                    var_dict[b] = b_bool
                    var_dict[c] = c_bool
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    # # F F T
                    var_dict[b] = (not b_bool)
                    var_dict[c] = (c_bool)
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    # # F T F
                    var_dict[b] = b_bool
                    var_dict[c] = (not c_bool)
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    else:
                        var_dict.pop(b)
                        var_dict.pop(c)
                        clause_list = [clause] + clause_list
                        return False
                    
            elif(assigned_b):
                if(var_dict[b] == b_bool):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    # # T F F
                    var_dict[a] = a_bool
                    var_dict[c] = (not c_bool)
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    # # T F T
                    var_dict[a] = a_bool
                    var_dict[c] = c_bool
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    # # F F T
                    var_dict[a] = (not a_bool)
                    var_dict[c] = c_bool
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    else:
                        var_dict.pop(a)
                        var_dict.pop(c)
                        clause_list = [clause] + clause_list
                        return False
              
            elif(assigned_c):
                if(var_dict[c] == c_bool):
                    bv = rec_7_alg(clause_list,var_dict)
                    if(bv):
                        return True
                    else:
                        clause_list = [clause] + clause_list
                        return False
                else:
                    # # T F F
                    var_dict[a] = a_bool
                    var_dict[b] = (not b_bool)
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    # # F T F
                    var_dict[a] = (not a_bool)
                    var_dict[b] = b_bool
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return (bool_val,v_dict)
                    # # T T F
                    var_dict[a] = a_bool
                    var_dict[b] = b_bool
                    bool_val = rec_7_alg(clause_list,var_dict)
                    if(bool_val):
                        return True
                    else:
                        var_dict.pop(b)
                        var_dict.pop(a)
                        clause_list = [clause] + clause_list
                        return False

            else:
                # # T T T
                var_dict[a] = a_bool
                var_dict[b] = b_bool
                var_dict[c] = c_bool
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # T F F
                var_dict[a] = a_bool
                var_dict[b] = (not b_bool)
                var_dict[c] = (not c_bool)
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # T F T
                var_dict[a] = a_bool
                var_dict[b] = (not b_bool)
                var_dict[c] = c_bool
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # T T F
                var_dict[a] = a_bool
                var_dict[b] = b_bool
                var_dict[c] = (not c_bool)
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # F T T
                var_dict[a] = (not a_bool)
                var_dict[b] = b_bool
                var_dict[c] = c_bool
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # F F T
                var_dict[a] = (not a_bool)
                var_dict[b] = (not b_bool)
                var_dict[c] = c_bool
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                # # F T F
                var_dict[a] = (not a_bool)
                var_dict[b] = b_bool
                var_dict[c] = (not c_bool)
                bool_val = rec_7_alg(clause_list,var_dict)
                if(bool_val):
                    return True
                else:
                    var_dict.pop(a)
                    var_dict.pop(b)
                    var_dict.pop(c)
                    clause_list = [clause] + clause_list
                    return False
# # #  End of rec_7_alg function    


# # # # # # # # #
# # Page Two # #
# # # # # # # #

def reachable_vertices(g, first_vertex):
    visited = set()
    q = [first_vertex]
    while q:
        v = q.pop(0)
        if v not in visited:
            visited.add(v)
            q.extend(g[v] - visited)
    return visited

def is_invalid(vert,neg_vert,g):     
    conn_vert = reachable_vertices(g,vert)
    conn_neg_vert = reachable_vertices(g,neg_vert)
    vert_to_neg_vert = (neg_vert in conn_vert)
    neg_vert_to_vert = (vert in conn_neg_vert)
    
    if(vert_to_neg_vert and neg_vert_to_vert):
        return True
    else:
        return False
    
def find_2_SAT_sol(g,vars_to_solve):
    assigned_vars = set()
    var_value = dict()
    l = len(vars_to_solve)
    for var in vars_to_solve:
        if(len(assigned_vars) < l):
            if(var[0] == '-'):
                neg_var = var[1:]
            else:
                neg_var = "-"+var

            conn_vert = reachable_vertices(g,var)
            if(not (neg_var in conn_vert)):
                diff_conn_assigned = conn_vert - assigned_vars
                if(len(diff_conn_assigned) > 0):
                    for v in diff_conn_assigned:
                        if(v[0] == '-'):
                            neg_v = v[1:]
                        else:
                            neg_v = "-"+v

                        var_value[v] = True
                        var_value[neg_v] = False
                        assigned_vars.add(v)
                        assigned_vars.add(neg_v)
        else:
            break
                    
    return var_value
            

# # # # # # # # #
# # Page Three # #
# # # # # # # #
def eval_2_SAT(exp_list_input):
    graph = dict()
    var_sol = dict()
    
    if(len(exp_list_input) == 0):
        return (True,var_sol)
    
    for clse in exp_list_input:
        a = clse[0]
        b = clse[1]
        
        if(a[0] == '-'):
            not_a = a[1:]
        else:
            not_a = "-"+a
            
        if(b[0] == '-'):
            not_b = b[1:]
        else:
            not_b = "-"+b
        
        if(not_a not in graph):
            adj_list = set()
            adj_list.add(b)
            graph[not_a] = adj_list
        else:
            graph[not_a].add(b)
            
        if(not_b not in graph):
            adj_list = set()
            adj_list.add(a)
            graph[not_b] = adj_list
        else:
            graph[not_b].add(a)
            
        if(a not in graph):
            graph[a] = set()
        if(b not in graph):
            graph[b] = set()
            
    visited_keys = []
    var_for_sol = []
    for k in graph:
        if(k not in visited_keys):      
            if(k[0] == '-'):
                k1 = k[1:]
                k2 = k
            else:
                k1 = k
                k2 = "-"+k
            var_for_sol.append(k1)
            var_for_sol.append(k2)
            visited_keys.append(k1)
            visited_keys.append(k2)

        if(is_invalid(k1,k2,graph)):
            return (False,var_sol)
    
    sol = find_2_SAT_sol(graph,var_for_sol)
    return (True,sol)
        

# # # # # # # # #
# # Page Four # #
# # # # # # # #

# # function get_list_of_clauses to change string input into list of clauses
def get_list_of_clauses(inp_input):
    exp_list = []
    while(len(inp_input) > 0):
        if(inp_input[0] == ";"):
            if(len(inp_input) == 1):
                inp_input = []
            else:
                inp_input = inp_input[1:]
        else:
            exp = []
            empty_or_semicolon = False
            while(not empty_or_semicolon):
                c = inp_input[0]
                if(len(inp_input) == 1):
                    inp_input = []
                    empty_or_semicolon = True
                else:
                    inp_input = inp_input[1:]

                if(c == ";"):
                    empty_or_semicolon = True
                elif(c == ","):
                    exp.append("or")
                else:
                    if(len(exp) > 0):
                        prev_char = exp[len(exp)-1]
                        if(prev_char != "or"):
                            c = prev_char+c
                            exp[len(exp)-1] = c
                        else:
                            exp.append(c)
                    else:
                        exp.append(c)
            exp_list.append(exp)
    
    return (exp_list)

# # # # # # # # #
# # Page Five # #
# # # # # # # #

def get_var_list(clse_list):
    var_set = set()
    if(len(clse_list) == 0):
        return var_set
    else:
        for cl in clse_list:
            for ltrl in cl:
                if(ltrl[0] == '-'):
                    var_set.add(ltrl[1:])
                else:
                    var_set.add(ltrl)
        return var_set

# # # # # # # # #
# # Page Six # #
# # # # # # # #

# #  Remove 'or' from cluases
def remove_or(exp_list_input):
    list_without_or = []    
    for clause in exp_list_input:
        if len(clause) == 1:
            list_without_or.append([clause[0]])
        elif len(clause) == 3:
            list_without_or.append([clause[0],clause[2]])
        elif len(clause) == 5:
            list_without_or.append([clause[0],clause[2],clause[4]])

    result_list = []
    for cls in list_without_or:
        cls = list(set(cls))
        result_list.append(cls)
    lst = []
    for cl in result_list:
        if( cl not in lst):
            lst.append(cl)
    return lst

# # End of remove_or function

# # # # # # # # # #
# # Page Seven # #
# # # # # # # # #
# #  Get list of variable and corresponding value that can be assigned easily 
# # i.e. sigle variable or variable with other assigned values
def get_bool_for_pos_var(exp_list_input,v_dict):
    dict_result = v_dict
    if(len(exp_list_input) == 0):
        return (True,exp_list_input,v_dict)
    
    list_of_single_var = [clse for clse in exp_list_input if len(clse) == 1]
    list_of_two_var = [clse for clse in exp_list_input if len(clse) == 2]
    list_of_three_var = [clse for clse in exp_list_input if len(clse) == 3]
    
    for clse in list_of_single_var:
        var_name = clse[0]
        if(var_name[0] == '-'):
            bool_val = False
            var_name = var_name[1:]
        else:
            bool_val = True
        if(var_name in dict_result):
            if(dict_result[var_name] != bool_val):
                return (False,exp_list_input,dict_result)
        else:
            dict_result[var_name] = bool_val
            
    done_with_all = False    
    while(not done_with_all):
        done_with_all = True
        
        ## list of two variables
        clse_to_remove = []
        for clause in list_of_two_var:
            [a,b] = clause
            if(a[0] == '-'):
                a = a[1:]
                a_bool = False
            else:
                a_bool = True
                
            if(b[0] == '-'):
                b = b[1:]
                b_bool = False
            else:
                b_bool = True
                
            assigned_a = (a in dict_result)
            assigned_b = (b in dict_result)
            
            if(a == b):
                clse_to_remove.append(clause)
            elif(assigned_a and assigned_b):
                ## both first and second variables are assigned
                if((dict_result[a] == a_bool) or (dict_result[b] == b_bool)):
                    clse_to_remove.append(clause)
                else:
                    return (False,exp_list_input,dict_result)
            elif(assigned_a):
                ## first variable is assigned
                if(dict_result[a] == a_bool):
                    clse_to_remove.append(clause)
                else:
                    dict_result[b] = b_bool
                    clse_to_remove.append(clause)
                    done_with_all = False
            elif(assigned_b):
                ## second variable is assigned
                if(dict_result[b] == b_bool):
                    clse_to_remove.append(clause)
                else:
                    dict_result[a] = a_bool
                    clse_to_remove.append(clause)
                    done_with_all = False
                        
        for clse in clse_to_remove:
            list_of_two_var.remove(clse)
            
        ## list of three variables
        clse_to_remove = []
        for clause in list_of_three_var:
            [a,b,c] = clause
            if(a[0] == '-'):
                a = a[1:]
                a_bool = False
            else:
                a_bool = True
                
            if(b[0] == '-'):
                b = b[1:]
                b_bool = False
            else:
                b_bool = True
                
            if(c[0] == '-'):
                c = c[1:]
                c_bool = False
            else:
                c_bool = True
                
            assigned_a = (a in dict_result)
            assigned_b = (b in dict_result)
            assigned_c = (c in dict_result)
            
            if((a == b) or (a == c) or (b == c)):
                clse_to_remove.append(clause)
            elif(assigned_a and assigned_b and assigned_c):
                ## all three variables are assigned
                if((dict_result[a] == a_bool) or (dict_result[b] == b_bool) or (dict_result[c] == c_bool)):
                    clse_to_remove.append(clause)
                else:
                    return (False,exp_list_input,dict_result)
            elif(assigned_a and assigned_b):
                ## first and second variables are assigned
                if((dict_result[a] == a_bool) or (dict_result[b] == b_bool)):
                    clse_to_remove.append(clause)
                else:
                    dict_result[c] = c_bool
                    clse_to_remove.append(clause)
                    done_with_all = False
            elif(assigned_a and assigned_c):
                ## first and last variables are assigned
                if((dict_result[a] == a_bool) or (dict_result[c] == c_bool)):
                    clse_to_remove.append(clause)
                else:
                    dict_result[b] = b_bool
                    clse_to_remove.append(clause)
                    done_with_all = False
            elif(assigned_b and assigned_c):
                ## second and last variables are assigned
                if((dict_result[b] == b_bool) or (dict_result[c] == c_bool)):
                    clse_to_remove.append(clause)
                else:
                    dict_result[a] = a_bool
                    clse_to_remove.append(clause)
                    done_with_all = False
            elif(assigned_a):
                # only first variable is assigned
                if(dict_result[a] == a_bool):
                    clse_to_remove.append(clause)
                else:
                    new_clause = [clause[1],clause[2]]
                    list_of_two_var.append(new_clause)
                    clse_to_remove.append(clause)
            elif(assigned_b):
                ## only second variable is assigned
                if(dict_result[b] == b_bool):
                    clse_to_remove.append(clause)
                else:
                    new_clause = [clause[0],clause[2]]
                    list_of_two_var.append(new_clause)
                    clse_to_remove.append(clause)
            elif(assigned_c):
                ## only third variable is assigned
                if(dict_result[c] == c_bool):
                    clse_to_remove.append(clause)
                else:
                    new_clause = [clause[0],clause[1]]
                    list_of_two_var.append(new_clause)
                    clse_to_remove.append(clause)    
        for clse in clse_to_remove:
            list_of_three_var.remove(clse)
            
        if(done_with_all):
            combined_list = list_of_two_var + list_of_three_var
            (new_val_assigned,dict_result) = assign_pure_literals(combined_list,dict_result)
            done_with_all =  (not new_val_assigned)
                
    exp_list_result = list_of_two_var + list_of_three_var
    return (True,exp_list_result,dict_result)
# # End of get_bool_for_pos_var  function

# # # Helper function for get_bool_for_pos_var to find pure literal in formula
def assign_pure_literals(formula,var_dictionary):
    var_set = set()
    var_to_assign = set()
    for cl in formula:
        for ele in cl:
            var_set.add(ele)
    while(len(var_set) > 0):
        v_name = var_set.pop()
        if(v_name[0] == '-'):
            neg_v_name = v_name[1:]
        else:
            neg_v_name = '-'+v_name
            
        if(neg_v_name not in var_set):
            var_to_assign.add(v_name)
        else:
            var_set.remove(neg_v_name)
    if(len(var_to_assign) > 0):
        for vr in var_to_assign:
            if(vr[0] == '-'):
                var_dictionary[vr[1:]] = False
            else:
                var_dictionary[vr] = True
        return (True,var_dictionary)
    else:
        return (False,var_dictionary)
# # end of function assign_pure_literals


# # # # # # # # # #
# # Page Eight # #
# # # # # # # # #

# # create list of partial assignment from given input which is in string format for Part Two
def get_list_of_part_assign(input_string):
    input_string = input_string.strip()
    var_dict = dict()
    if(len(input_string) > 0):
        input_string = input_string.split(',')
        for var in input_string:
            [key,value] = var.split('=')
            if(key[0] == '-'):
                if(value == "T"):
                    var_dict[key[1:]] = False
                elif(value == "F"):
                    var_dict[key[1:]] = True
                    
            if(value == "T"):
                var_dict[key] = True
            elif(value == "F"):
                var_dict[key] = False
            
    return var_dict
# # End of get_list_of_part_assign function

# # # # # # # # #
# # Page Nine # #
# # # # # # # # #

# # Check whether given input is a 3-SAT or 2-SAT formula
def is_2_SAT(lst):
    if(len(lst) == 0):
        return True
    else:
        not_3_sat = True
        for c in lst:
            if(len(c) == 3):
                not_3_sat = False
                return False
        return not_3_sat
# # End of is_2_SAT function

#### Testing
outputs = {'part1_one': 'no', 'part1_two':"", 'part2_one':'no', 'part2_two': "", 'part3_one':'no', 'part3_two': ""}   # for the testing purpose


# # # # # # # # # # # # #
# # Page For Part One # #
# # # # # # # # # # # # #

# #  Handle input and evaluate formula for part one
exp_list = get_list_of_clauses(inp)  
  
exp_list = remove_or(exp_list)

init_var_set = get_var_list(exp_list)

##### Testing
formula1 = exp_list.copy()
variables1 = init_var_set.copy()

(is_satisfiable,exp_list,var_val_dict) = get_bool_for_pos_var(list(exp_list),dict())

if(is_satisfiable):
    
    (solved,val_of_var) = eval_2_SAT(list(exp_list))
    
    if(solved):
        keys = list(var_val_dict.keys())
        for k in keys:
            if(k[0] == '-'):
                if(k[1:] not in var_val_dict):
                    var_val_dict[k[1:]] = (not var_val_dict[k])
                    var_val_dict.pop(k)
                else:
                    var_val_dict.pop(k)
                        
        keys = list(val_of_var.keys())
        for k in keys:
            if(k[0] == '-'):
                if(k[1:] not in val_of_var):
                    val_of_var[k[1:]] = (not val_of_var[k])
                    val_of_var.pop(k)
                else:
                    val_of_var.pop(k)
        
        
        final_var_set1 = set(var_val_dict.keys())
        final_var_set2 = set(val_of_var.keys())
        final_var_set = (final_var_set1|final_var_set2)
        ann_var_set = init_var_set - final_var_set
        if(len(ann_var_set) > 0):
            for ann_var in ann_var_set:
                var_val_dict[ann_var] = True
        for key in var_val_dict:
            val_of_var[key] = var_val_dict[key]
        var_val_dict = dict()
       
        print("yes\n")
        my_string = ""
        
        for key,value in val_of_var.items():
            if(key[0] != '-'):
                my_string = my_string+key+"="
                if(value):
                    my_string = my_string+"T,"
                else:
                    my_string = my_string+"F,"
                    
        my_string = my_string[0:len(my_string)-1]
        print(my_string+"\n")
        outputs['part1_one'] = "yes"
        outputs['part1_two'] = my_string
        
    else:
        print("no\n")
                
    
else:
    print("no\n")
    
print("$\n")

# # End of evaluation for part one


# # # # # # # # # # # # #
# # Page For Part Two # #
# # # # # # # # # # # # #

# # # Handle input and evaluate for Part Two

var_val_dict2 = get_list_of_part_assign(part_assign)

exp_list2 = get_list_of_clauses(inp2)
  
exp_list2 = remove_or(exp_list2)

# # f2 = exp_list2.copy()

init_var_set = get_var_list(exp_list2)

#### Testing
formula2 = exp_list2.copy()
variables2 = init_var_set.copy()


(is_satisfiable2,exp_list2,var_val_dict2) = get_bool_for_pos_var(list(exp_list2),var_val_dict2.copy())

if(is_satisfiable2):
    if(is_2_SAT(exp_list2)):
        (solved2,val_of_var2) = eval_2_SAT(list(exp_list2))
    
        if(solved2):
            keys = list(var_val_dict2.keys())
            for k in keys:
                if(k[0] == '-'):
                    if(k[1:] not in var_val_dict2):
                        var_val_dict2[k[1:]] = (not var_val_dict2[k])
                        var_val_dict2.pop(k)
                    else:
                        var_val_dict2.pop(k)
                            
            keys = list(val_of_var2.keys())
            for k in keys:
                if(k[0] == '-'):
                    if(k[1:] not in val_of_var2):
                        val_of_var2[k[1:]] = (not val_of_var2[k])
                        val_of_var2.pop(k)
                    else:
                        val_of_var2.pop(k)
                        
            final_var_set1 = set(var_val_dict2.keys())
            final_var_set2 = set(val_of_var2.keys())
            final_var_set = (final_var_set1|final_var_set2)
            ann_var_set = init_var_set - final_var_set
            if(len(ann_var_set) > 0):
                for ann_var in ann_var_set:
                    var_val_dict2[ann_var] = True 
            for key in var_val_dict2:
                val_of_var2[key] = var_val_dict2[key]
            var_val_dict2 = dict()
            
            print("yes\n")
            my_string = ""
            for key,value in val_of_var2.items():
                if(key[0] != '-'):
                    my_string = my_string+key+"="
                    if(value):
                        my_string = my_string+"T,"
                    else:
                        my_string = my_string+"F,"

            my_string = my_string[0:len(my_string)-1]
            print(my_string+"\n")
            outputs['part2_one'] = "yes"
            outputs['part2_two'] = my_string

        else:
            print("no\n")
    else:
        total_no = 0
        total_yes = 0
        for clse in exp_list2:
            len_of_clse = len(clse)
            num_of_false = 0
            num_of_true = 0
            for var in clse:
                bool_val = True
                if(var[0] == '-'):
                    bool_val = False
                    var = var[1:]
                    
                if(var in var_val_dict2):
                    if(var_val_dict2[var] == bool_val):
                        num_of_true = num_of_true + 1
                    else:
                        num_of_false = num_of_false + 1
                        
            if(num_of_false == len_of_clse):
                total_no = 1
            elif(num_of_true > 0):
                total_yes = total_yes + 1
        if(total_no > 0):
            print("no\n")
        elif(total_yes == len(exp_list2)):
            keys = list(var_val_dict2.keys())
            for k in keys:
                if(k[0] == '-'):
                    if(k[1:] not in var_val_dict2):
                        var_val_dict2[k[1:]] = (not var_val_dict2[k])
                        var_val_dict2.pop(k)
                    else:
                        var_val_dict2.pop(k)

            final_var_set1 = set(var_val_dict2.keys())
            final_var_set2 = set()
            final_var_set = (final_var_set1|final_var_set2)
            ann_var_set = init_var_set - final_var_set
            if(len(ann_var_set) > 0):
                for ann_var in ann_var_set:
                    var_val_dict2[ann_var] = True 
                    
            print("yes\n")
            my_string = ""
            for key,value in var_val_dict2.items():
                my_string = my_string+key+"="
                if(value):
                    my_string = my_string+"T,"
                else:
                    my_string = my_string+"F,"
            my_string = my_string[0:len(my_string)-1]
            print(my_string+"\n")
            outputs['part2_one'] = "yes"
            outputs['part2_two'] = my_string
            
        else:
            final_var_set = get_var_list(exp_list2)
            ann_var_set = init_var_set - final_var_set
            for ann_var in ann_var_set:
                if(ann_var not in var_val_dict2):
                    var_val_dict2[ann_var] = True

            print("maybe\n")
            my_string = ""
            for key,value in var_val_dict2.items():
                my_string = my_string+key+"="
                if(value):
                    my_string = my_string+"T,"
                else:
                    my_string = my_string+"F,"
            my_string = my_string[0:len(my_string)-1]
            print(my_string+"\n")
            outputs['part2_one'] = "maybe"
            outputs['part2_two'] = my_string
            
else:
    print("no\n")


print("$\n")


# # # # # # # # # # # # # #
# # Page For Part Three # #
# # # # # # # # # # # # # #

exp_list3 = get_list_of_clauses(inp3)  
  
exp_list3 = remove_or(exp_list3)

init_var_set = get_var_list(exp_list3)

############ Testing
formula3 = exp_list3.copy()
variables3 = init_var_set.copy()

#####print(len(exp_list3))
#####print(len(init_var_set))

(is_satisfiable3,exp_list3,var_val_dict3) = get_bool_for_pos_var(list(exp_list3),dict())

if(is_satisfiable3):
    solved = rec_7_alg(list(exp_list3),var_val_dict3.copy())
    val_of_var3 = dict()
    if(solved):
        keys = list(var_val_dict3.keys())
        for k in keys:
            if(k[0] == '-'):
                if(k[1:] not in var_val_dict3):
                    var_val_dict3[k[1:]] = (not var_val_dict3[k])
                    var_val_dict3.pop(k)
                else:
                    var_val_dict3.pop(k)
        keys = list(val_of_var3.keys())
        for k in keys:
            if(k[0] == '-'):
                if(k[1:] not in val_of_var3):
                    val_of_var3[k[1:]] = (not val_of_var3[k])
                    val_of_var3.pop(k)
                else:
                    val_of_var3.pop(k)
        
        
        final_var_set1 = set(var_val_dict3.keys())
        final_var_set2 = set(val_of_var3.keys())
        final_var_set = (final_var_set1|final_var_set2)
        ann_var_set = init_var_set - final_var_set
        
        if(len(ann_var_set) > 0):
            for ann_var in ann_var_set:
                var_val_dict3[ann_var] = True
        for key in val_of_var3:
            var_val_dict3[key] = val_of_var3[key]
        val_of_var3 = dict()

        print("yes\n")
        my_string = ""
        for key,value in var_val_dict3.items():
            my_string = my_string+key+"="
            if(value):
                my_string = my_string+"T,"
            else:
                my_string = my_string+"F,"            
        my_string = my_string[0:len(my_string)-1]
        print(my_string+"\n")
        outputs['part3_one'] = "yes"
        outputs['part3_two'] = my_string
    else:
        print("no\n")
                
    
else:
    print("no\n")
    
# # End of evaluation for part three



############################
####  Code for testing  ####
############################


# print("Testing the outputs\n_____________________")
# print("Part One\n")
# if(outputs['part1_one'] == 'yes'):
#     print("For Yes")
#     s = outputs['part1_two']
#     print(s)
#     assigned_variables = (get_list_of_part_assign(s))
#     if(len(variables1) != len(assigned_variables.keys())):
#         print("some of the variables are not assigned yet")
#         raise ValueError('Not all variables are assigned yet')
#     else:
#         result = "the solution works on part one"
#         for clause in formula1:
#             b = False
#             for var in clause:
#                 if(var[0] == '-'):
#                     if(assigned_variables[var[1:]] == False):
#                         b = True
#                 else:
#                     if(assigned_variables[var] == True):
#                         b = True
#             if(not b):
#                 print("does not work for this : ",clause)
#                 result = result = "solution does not work on part one"
#                 raise ValueError('A very specific bad thing happened. cluase is ',clause)
#         print(result)
            
                    

# print("\nPart Two\n")
# if(outputs['part2_one'] == 'yes'):
#     print("For Yes")
#     s = outputs['part2_two']
#     print(s)
#     assigned_variables = (get_list_of_part_assign(s))
#     if(len(variables2) != len(assigned_variables.keys())):
#         print("some of the variables are not assigned yet")
#         raise ValueError('Not all variables are assigned yet')
#     else:
#         result = "the solution works on part two"
#         for clause in formula2:
#             b = False
#             for var in clause:
#                 if(var[0] == '-'):
#                     if(assigned_variables[var[1:]] == False):
#                         b = True
#                 else:
#                     if(assigned_variables[var] == True):
#                         b = True
#             if(not b):
#                 print("does not work for this : ",clause)
#                 result = result = "solution does not work on part two"
#                 raise ValueError('A very specific bad thing happened. cluase is ',clause)
#         print(result)
        
        
        

# if(outputs['part2_one'] == 'maybe'):
#     print("For Maybe")
#     s = outputs['part2_two']
#     assigned_variables = (get_list_of_part_assign(s))
#     if(len(variables2) != len(assigned_variables.keys())):
#         print("some of the variables are not assigned yet")
        
# print("\nPart Three\n")
# if(outputs['part3_one'] == 'yes'):
#     print("For Yes")
#     s = outputs['part3_two']
#     print(s)
#     assigned_variables = (get_list_of_part_assign(s))
#     if(len(variables3) != len(assigned_variables.keys())):
#         print("some of the variables are not assigned yet")
#         raise ValueError('Not all variables are assigned yet')
#     else:
#         result = "the solution works on part three"
#         for clause in formula3:
#             b = False
#             for var in clause:
#                 if(var[0] == '-'):
#                     if(assigned_variables[var[1:]] == False):
#                         b = True
#                 else:
#                     if(assigned_variables[var] == True):
#                         b = True
#             if(not b):
#                 print("does not work for this : ",clause)
#                 result = result = "solution does not work on part three"
#                 raise ValueError('A very specific bad thing happened. cluase is ',clause)
#         print(result)
# else:
#     print("No")
```
